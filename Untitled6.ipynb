2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define values for a discrete uniform distribution (e.g., rolling a fair die)
values = np.arange(1, 7)
probabilities = np.ones_like(values, dtype=float) / len(values)

# Plot the PMF without the use_line_collection argument
plt.stem(values, probabilities)
plt.title("Discrete Uniform Distribution - PMF")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True)
plt.show()

```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(p, x):
    return p**x * (1 - p)**(1 - x)

print("PDF of Bernoulli(p=0.7, x=1):", bernoulli_pdf(0.7, 1))
```

13. Generate random variables and calculate their corresponding probabilities using the binomial distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success

# Generate random variable outcomes
x = np.arange(0, n + 1)
probs = binom.pmf(x, n, p)

# Plot the PMF
plt.stem(x, probs)  # Removed use_line_collection
plt.title("Binomial Distribution PMF (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid(True)
plt.show()

# Print the values
for i in range(len(x)):
    print(f"P(X = {x[i]}) = {probs[i]:.4f}")
```

15. Implement hypothesis testing using Z-statistics for a sample dataset

```python
sample = np.random.normal(50, 10, 30)
z_stat = (np.mean(sample) - 50) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
p_val = 2 * (1 - norm.cdf(abs(z_stat)))
print("Z-statistic:", z_stat, "P-value:", p_val)
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python
import scipy.stats as stats
ci = stats.norm.interval(0.95, loc=np.mean(sample), scale=stats.sem(sample))
print("95% Confidence Interval:", ci)
```
